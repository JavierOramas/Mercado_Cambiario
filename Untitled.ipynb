```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('data.json') as f:    
    data = json.load(f)
df = pd.DataFrame(data)
```

```python
from random import random


class Agent:
    def __init__(self, type):
        # -1 for desperate, 0 for neutral, 1 for greedy
        self.type = type
        # value of despair (either positive or negative) in the agent
        self.delta = np.random.choice([0.1, 0.3, 0.5, 0.7])
    def run(self, value):
        return value + self.delta*self.type
```

```python
# take first 300 rows from df
price = df.iloc[3:300]
prices = df['median'].tolist()
price_mean = []
# 3 day window for mean
for i in range(3,300):
        price_mean.append(int((prices[i-1]+prices[i-2]+prices[i-3]+prices[i])/4))
price['mean'] = price_mean
price['mode'] = price['median']

price = price[['mean', 'mode', 'median']]
price
```

```python
from scipy.stats import skewnorm
from numpy import random

normal = lambda n : random.normal(loc=1,scale=10, size=n) 
uniform = lambda n : random.uniform(low=-1, high=10, size=n)
exponential = lambda n : random.exponential(scale=10,size=n)
poisson = lambda n : random.poisson(lam=2, size=n) 
skewnor  = lambda n : skewnorm.rvs(2, loc=5, scale=10, size=n)

distributions = {
    "normal" : normal,
    "skewnorm" : skewnor,
    "uniform": uniform,
    "exponential": exponential,
    "poisson": poisson, 
}
```

```python
def generate_agents(dist):
    x = distributions[dist](500)
    agents = []
    for i in x:
        if i < 0:
            type = -1
        if i > 10:
            type = 1
        else:
            type = 0
        agents.append(Agent(type))
    return agents
```

```python
from scipy.stats import mode

def run(days, dist, prices):
    
    agents = generate_agents(dist)
    

    new_prices = prices.copy()
    
    for i in range(days):

        day_prices_mean = []
        day_prices_median = []
        day_prices_mode = []

        for agent in agents:
            day_prices_mean.append(agent.run(new_prices['mean'].iloc[-1]))
            day_prices_mode.append(agent.run(new_prices['mode'].iloc[-1]))
            day_prices_median.append(agent.run(new_prices['median'].iloc[-1]))

        today_mean = np.mean(day_prices_mean)
        today_median = np.median(day_prices_median)
        today_mode = int(mode(day_prices_mode)[0])

        # append today's prices to the dataframe
        new_prices = new_prices.append({'mean': today_mean, 'median': today_median, 'mode': today_mode}, ignore_index=True)
    # dump prices to text file
    with open("file.txt", "a") as file:
        file.write(str(new_prices.tail()))
        file.write("\n")
```

```python
with open("file.txt", "w") as file:
    file.write("starting simulations\n")
for dist in distributions.keys():
    with open("file.txt", "a") as file:
        file.write(f"--------------------- {dist} ----------------------\n")
    print(dist)
    for i in range(30):
        with open("file.txt", "a") as file:
            file.write(f"simulation {i+1}/30\n")
        print(f"simulation {i+1}/30")
        run(days = 700, dist=dist, prices=price)
```
